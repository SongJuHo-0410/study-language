## 1.     .( 2       ) 2   
 


```python
age = int(input('  :'))

if 8 <= age <= 19:
    print('%d 300' % age)
elif 20 <= age <= 64:
    print('%d 500' % age)
else:
    print('%d 0' % age)
```

##    ,   .(:     !)  3  


```python
num1 = int(input('   :'))
num2 = int(input('   :'))
operator = (input('(+ - * /)' ))

if operator == '+':
    print('%d %s %d = %d' % (num1,operator,num2,num1+num2))
elif operator == '-':
    print('%d %s %d = %d' % (num1,operator,num2,num1-num2))
elif operator == '*':
    print('%d %s %d = %d' % (num1,operator,num2,num1*num2))
elif operator == '/':
    print('%d %s %d = %d' % (num1,operator,num2,num1/num2))
```
