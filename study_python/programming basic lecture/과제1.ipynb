## 1.아래 표에맞는 조건문을 이용하여 프로그램을 완성하시오.(단 2개의 조건문만 이용하고 나이는 정수를 키보드로 입력받아 사용) 2점   
 


```python
age = int(input('당신의 나이를 입력하세요:'))

if 8 <= age <= 19:
    print('%d 300' % age)
elif 20 <= age <= 64:
    print('%d 500' % age)
else:
    print('%d 0' % age)
```

## 두 정수를 키보드로 입력받고, 연산의 결과를 출력한다.(주의: 나눗셈 사칙 연산은 정수로 출력!)  3점  


```python
num1 = int(input('첫 번째 수 입력:'))
num2 = int(input('두 번째 수 입력:'))
operator = (input('(+ - * /)' ))

if operator == '+':
    print('%d %s %d = %d' % (num1,operator,num2,num1+num2))
elif operator == '-':
    print('%d %s %d = %d' % (num1,operator,num2,num1-num2))
elif operator == '*':
    print('%d %s %d = %d' % (num1,operator,num2,num1*num2))
elif operator == '/':
    print('%d %s %d = %d' % (num1,operator,num2,num1/num2))
```
