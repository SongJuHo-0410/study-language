##    (factorial)   (for )  2  
  n! = n * (n-1) * (n-2) * * 1


```python
n = int(input(" : "))
total = 1

for i in range(0,n,1):
    total *= (n-i)
    
print(total)
```

 ## for  1~100   3     . 3  


```python
hap = 0
for i in range(1,101,1):
    if i % 3 == 0:
        print(i, end =" ")
        hap += i
print("")
print("--------------------------------------")
print(hap)
```
