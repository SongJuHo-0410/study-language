## 하나의 정수를 입력받아 팩토리얼(factorial)을 구하는 프로그램을 작성하시오(for문 사용)  2점  
  n! = n * (n-1) * (n-2) * ………* 1


```python
n = int(input("숫자를 입력: "))
total = 1

for i in range(0,n,1):
    total *= (n-i)
    
print(total)
```

 ## for문을 이용하여 1~100까지의 정수 중에서 3의 배수의 합계를 구하는 프로그램을 완성하시오. 3점  


```python
hap = 0
for i in range(1,101,1):
    if i % 3 == 0:
        print(i, end =" ")
        hap += i
print("")
print("--------------------------------------")
print(hap)
```
